```python
import pandas as pd
```

```python
df = pd.read_csv('data/lynx_ppl.csv', header=None)
df.drop(columns=4, inplace=True)
```

```python
df.columns = ['athleteNumber', 'lastName', 'firstName', 'team', 'sex', 'eventNumber']
```

```python
def get_all_events(group):
    eventNumbers = group.eventNumber.tolist()
    strEventNumbers = [str(num) for num in eventNumbers]
    group['eventNumbers'] = ','.join(strEventNumbers)
    return group
```

```python
df = df.groupby(by='athleteNumber').apply(get_all_events)
df.drop(columns='eventNumber', inplace=True)
df.drop_duplicates(subset='athleteNumber', inplace=True)
df.head()
```
